```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import modules.helpers as hlp
import numpy as np
import random
import cv2
```

```python
training_images = "day_night_images/training/"
test_images = "day_night_images/test/"
```

```python
IMAGE_LIST = hlp.load_dataset(training_images)
```

```python
STANDARDIZED_IMAGES = hlp.standardize(IMAGE_LIST)
```

```python
def average_brightness(image2):
    selectedimage = STANDARDIZED_IMAGES[image2][0]
    selectedimagelabel = STANDARDIZED_IMAGES[image2][1]
    hsvimage = cv2.cvtColor(selectedimage, cv2.COLOR_RGB2HSV)
    average_brightness = np.sum(hsvimage[:, :, 2])/(1100*600)
    return average_brightness
```

```python
image = int(input())
plt.imshow(STANDARDIZED_IMAGES[image][0])
average_brightness(image)
```

```python
def estimator(image1):
    if average_brightness(image1) > 125:
        return 1
    else:
        return 0
```

```python
estimator(119)
```

```python
TEST_IMAGE_LIST = hlp.load_dataset(test_images)

STANDARDIZED_TEST_LIST = hlp.standardize(TEST_IMAGE_LIST)

```

```python
def accuracy(image_list):
    wrong = []
    correct = 0
    for i in range(0, len(image_list)):
        if estimator(i) == image_list[i][1]:
            correct+=1
        else:
            wrong.append(i)
    return (correct/len(STANDARDIZED_TEST_LIST))*100
```
