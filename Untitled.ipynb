## Performing Edge Detection

```python
#convert to grayscale
imone_1_gray = cv2.cvtColor(imone_1, cv2.COLOR_RGB2GRAY)

#defining canny parameters
low_threshold = 80
high_threshold = 90

imone_edges = cv2.Canny(imone_1_gray, low_threshold, high_threshold)
plt.imshow(imone_edges)
```
